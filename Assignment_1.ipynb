```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
ipl_data = pd.read_csv('/content/iris dataset.csv')

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

# Pairplot
plt.figure(figsize=(10, 8))
sns.pairplot(df, hue='target', markers=['o', 's', 'D'])
plt.show()

# Barplot
plt.figure(figsize=(8, 6))
sns.countplot(x='target', data=df)
plt.title('Target Count')
plt.show()

# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['sepal length (cm)'], bins=10, kde=True)
plt.title('Sepal Length Distribution')
plt.show()

# Distribution Plot
plt.figure(figsize=(8, 6))
sns.distplot(df['sepal width (cm)'], bins=10, kde=True)
plt.title('Sepal Width Distribution')
plt.show()

# Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='petal length (cm)', data=df)
plt.title('Petal Length Boxplot')
plt.show()

# Heatmap
plt.figure(figsize=(10, 8))
corr_matrix = df.drop('target', axis=1).corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()

# Scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='target', data=df)
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
ipl_data = pd.read_csv('/content/ipl data set.csv')
print(ipl_data.head())
print(ipl_data.info())
print(ipl_data.describe())
print(ipl_data.isnull().sum())
plt.hist(ipl_data['result'])
plt.xlabel('Runs Scored')
plt.ylabel('Frequency')
plt.title('Distribution of Runs Scored by Each Team')
plt.show()

top_teams = ipl_data['team'].value_counts().head(10)
plt.bar(top_teams.index, top_teams.values)
plt.xlabel('Team')
plt.ylabel('Wins')
plt.title('Top 10 Teams with Most Wins')
plt.show()
sns.heatmap(ipl_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between Columns')
plt.show()

sns.boxplot(x='team', y='runs_scored', data=ipl_data)
plt.title('Boxplot of Runs Scored by Each Team')
plt.show()
```
